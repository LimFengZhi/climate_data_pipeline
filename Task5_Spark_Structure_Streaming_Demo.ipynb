# Task 5 Spark Structure Streaming

```python
from de_classes.streamer_classes.climate_risk_scoring_streamer import ClimateRiskScoringStreamer
from de_classes.utility_classes.spark_manager import SparkManager
from dotenv import load_dotenv
from datetime import datetime
import os
load_dotenv()
```

```python
config_spark = {
                "spark.sql.adaptive.enabled": "false",
                "spark.sql.streaming.stateStore.maintenanceInterval" : "60s",
                "spark.sql.adaptive.coalescePartitions.enabled" : "false",
                "spark.sql.streaming.stateStore.maintenanceInterval": "60s",
                "spark.sql.streaming.stateStore.minDeltasForSnapshot": "1",
                "spark.sql.streaming.checkpointLocation.deleteOnStop": "false",
                "spark.serializer": "org.apache.spark.serializer.KryoSerializer"
               }
PROCESSED_WEATHER_PATH = os.getenv('CLEANED_WEATHER')
```

```python
with SparkManager(app_name = "Stream Spark", configs= config_spark) as manager:
    stream_spark = manager.spark
    stream_spark._jvm.org.apache.log4j.LogManager.getLogger("org.apache.spark.sql.execution.streaming.state.HDFSBackedStateStoreProvider").setLevel(stream_spark._jvm.org.apache.log4j.Level.ERROR)
    streamer = ClimateRiskScoringStreamer(spark = stream_spark, stream_file_path = PROCESSED_WEATHER_PATH)
    streamer.process_stream()
```
